# Marketing Funnel & Conversion Performance Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('marketing_funnel_data.csv')
df.head()
```

```python
# Calculate conversion rates
df['Lead_to_Qualified_%'] = (df['Qualified_Leads'] / df['Leads']) * 100
df['Qualified_to_Demo_%'] = (df['Demo_Booked'] / df['Qualified_Leads']) * 100
df['Demo_to_Customer_%'] = (df['Customers'] / df['Demo_Booked']) * 100

print(df)
```

```python
# Revenue visualization
plt.figure(figsize=(8,5))
plt.bar(df['Channel'], df['Revenue_USD'])
plt.xticks(rotation=20)
plt.ylabel('Revenue')
plt.title('Revenue by Marketing Channel')
plt.show()
```
